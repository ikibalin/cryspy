```python
f_name = "structure.cif"
obj = cryspy.load_file(f_name)
```

```python
obj.crystal_dyal
```

```python
?obj.crystal_dyal.calc_hkl
```

```python
np_h, np_k, np_l, np_mult = obj.crystal_dyal.calc_hkl(0., 0.18)
```

```python
print("   H   K   L  Mult")
for h, k, l, mult in zip(np_h, np_k, np_l, np_mult):
    print(f"{h:4}{k:4}{l:4}  {mult:4}")
```
